```python
import pandas as pd

df = pd.read_csv("Cargo Volume - Train.csv")
df
```

```python
df.columns
```

### Split Activity Period to Year and Month

```python
df['Activity Period Year'] = df['Activity Period'].astype('str').str[0:4]
df['Activity Period Month'] = df['Activity Period'].astype('str').str[4:6]
df
```

### Using LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
for i in range(1, 13):
    if i==10:
        continue
    #print(i)
    le.fit(df.iloc[:, i])
    df_labeled_col = le.transform(df.iloc[:, i])
    col_label = "col_label_%d" % (i)
    df[col_label] = df_labeled_col
df
```

```python
le.fit(df['Cargo Aircraft Type'])
le.classes_
```

```python
df.columns
```

```python
df.drop(['Operating Airline','Operating Airline IATA Code','Published Airline','Published Airline IATA Code',
         'GEO Summary','GEO Region','Activity Type Code','Cargo Type Code','Cargo Aircraft Type',
         'Activity Period Year','Activity Period Month'], axis=1, inplace=True)
df
```

```python
import numpy as np

def train_test_split(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]

train_data, valid_data = train_test_split(df, 0.2)
print(len(train_data), "train instances and", len(valid_data), "valid instances")
```

```python
train_data
```

```python
train_y = train_data.iloc[:, 1]
valid_y = valid_data.iloc[:, 1]
valid_y
```

```python
valid_data
```

```python
train_activity_period = train_data['Activity Period']
train_activity_period
```

```python
valid_activity_period = valid_data['Activity Period']
valid_activity_period
```

```python
train_data.drop('Activity Period', axis=1, inplace=True)
valid_data.drop('Activity Period', axis=1, inplace=True)
train_data
```

```python
train_data.drop(['Cargo Metric TONS'], axis=1, inplace=True)
valid_data.drop(['Cargo Metric TONS'], axis=1, inplace=True)
valid_data
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(train_data, train_y)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

np.sqrt(mean_squared_error(valid_y, reg.predict(valid_data)))
```

### using StandScaler

```python
reg = LinearRegression()
reg.fit(train_transformed_data, train_y)
np.sqrt(mean_squared_error(valid_y, reg.predict(valid_transformed_data)))
```
